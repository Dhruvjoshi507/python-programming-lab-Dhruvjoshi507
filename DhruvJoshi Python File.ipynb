```python
import random

while True:
    print("\nRock-Paper-Scissors Game!")
    print("1 = Rock, 2 = Paper, 3 = Scissors")

    user = int(input("Enter your choice (1/2/3): "))
    computer = random.randint(1, 3)

    print(f"Computer chose: {computer}")

    if user == computer:
        print("It's a tie!")
    elif (user == 1 and computer == 3) or (user == 2 and computer == 1) or (user == 3 and computer == 2):
        print("You win! ")
    else:
        print("Computer wins! ")

    play_again = input("Do you want to play again? (yes/no): ").lower()
    if play_again != "yes":
        print("Thanks for playing!")
        break
```

```python
import random
import string

def generate_password(length=12):
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for _ in range(length))
    return password

def main():
    while True:
        length = int(input("Enter password length: "))
        print("Generated Password:", generate_password(length))
        again = input("Do you want to generate another password? (yes/no): ").lower()
        if again != "yes":
            print("Goodbye!")
            break

if __name__ == "__main__":
    main()
```

```python
def largest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

# User input
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))

print("The largest number is:", largest(a, b, c))
```

```python
# Base class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Derived class
class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary

    def display(self):
        super().display()
        print(f"Salary: {self.salary}")

    # Operator overloading for +
    def __add__(self, other):
        return self.salary + other.salary


# User input for first employee
name1 = input("Enter first employee's name: ")
age1 = int(input("Enter first employee's age: "))
salary1 = float(input("Enter first employee's salary: "))

# User input for second employee
name2 = input("Enter second employee's name: ")
age2 = int(input("Enter second employee's age: "))
salary2 = float(input("Enter second employee's salary: "))

# Creating employee objects
emp1 = Employee(name1, age1, salary1)
emp2 = Employee(name2, age2, salary2)

# Display details
print("\nEmployee 1 Details:")
emp1.display()

print("\nEmployee 2 Details:")
emp2.display()

# Operator overloading demonstration
print("\nTotal Salary:", emp1 + emp2)
```

```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    result = num1 / num2
    print("Result:", result)

except ValueError:
    print("Error: Please enter valid integers only.")

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

except Exception as e:
    print("An unexpected error occurred:", e)

else:
    print("Division successful!")

finally:
    print("Program execution completed.")
```
